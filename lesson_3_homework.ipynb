```python
import pandas as pd
import numpy as np
```

```python
energy = pd.read_excel("Energy Indicators.xls", skiprows=16, skipfooter=38)
energy = energy.iloc[:, 2:]
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capital', '% Renewable']
energy = energy.replace(r'.*\..*', np.nan, regex = True)
energy.loc[0, 'Energy Supply'] = 'Gigajoules'
energy.loc[1:, 'Energy Supply'] = energy.loc[1:, 'Energy Supply'] * 1000000
energy['Country'] = energy['Country'].str.replace(r'\d+', '', regex=True)
energy['Country'] = energy['Country'].replace({
    "Republic of Korea":  "South Korea",

    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
})
energy




```

```python
gdp = pd.read_csv('data.csv', skiprows = 4)
gdp['Country Name'] = gdp['Country Name'].replace({
    
    "Korea, Rep." : "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"

})
gdp.rename(columns={'Country Name': 'Country'}, inplace=True)

gdp_filtered = gdp[["Country"] + [str(year) for year in range(2006, 2016)]]

gdp_filtered
```

```python
ScimEn = pd.read_excel("scim.xlsx")
ScimEn_filtered = ScimEn[ScimEn['Rank'] <= 15]
ScimEn_filtered
#
```

```python
def answer_one():
    conclusion = energy.merge(ScimEn_filtered).merge(gdp_filtered)
    conclusion.set_index('Country', inplace=True)
    conclusion = conclusion.sort_values(by="Rank")
    return conclusion
answer_one()
```

```python
df = answer_one()
def answer_two():
    needed_df = df.loc[:,'2006':'2015']
    needed_df['avgGDP'] = needed_df.mean(axis=1,skipna=True)
    needed_df = needed_df[['avgGDP']]
    return needed_df.sort_values(by='avgGDP', ascending=False)
answer_two()
```

```python
def answer_three():
    row = answer_two().index[5]
    changes = df.loc[row, '2015'] - df.loc[row, '2006']
    return changes.item()
   
answer_three()

    
```

```python
def answer_four():
    df['Ratio'] = df['Self-citations']/df['Citations']
    max_ratio = df['Ratio'].max()
    country_with_max_value = df[df['Ratio'] == max_ratio].index[0]
    return(country_with_max_value, max_ratio.item())
answer_four()
```

```python
def answer_five():
    df['Population'] = df['Energy Supply']/df['Energy Supply per Capital']
   
    needed_answer = df['Population'].sort_values(ascending=False).index[2]
    return needed_answer
answer_five()
df
```

```python
def answer_six():
    df['Documents per person'] = df['Citable documents']/ df['Population']
    corr = df['Documents per person'].corr(df['Energy Supply per Capital'])
    return corr.item()
answer_six()
```

```python
def answer_seven():
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    data = pd.DataFrame({"Continent" : ['Asia', 'Australia', 'Europe', 
                        'North America', 'South America'], "size" : [0] * 5,
                          'sum' : [0.0] * 5, 'mean' : [0.0] * 5, 'std' : [0.0] * 5})
    data = data.set_index('Continent')
    sum_dict = {'Asia': [],
                 'North America' : [],
                 'Europe' : [],
                 'Australia': [],
                 'South America' : []
                 }
    std_dict = {'2006': 0,
                 '2007' : 0,
                 '2008' : 0,
                 '2009': 0,
                 '2010' : 0,
                 '2011':0,
                 '2012':0,
                 '2013':0,
                 '2014':0,
                 '2015':0,
                 
                 
                 }
    for country, continent in ContinentDict.items():
        
        if(country in df.index):
            data.loc[continent, 'size']+=1
            data.loc[continent,'sum'] += (df.loc[country, 'Population'])
            sum_dict[continent] .append(df.loc[country, 'Population'])
            for key in std_dict:
                std_dict[key] = (df.loc[country, key])
            mas = []
            for value in std_dict.values():
                mas.append(value)
            check_std  = np.std(mas)
            if(data.loc[continent, 'std'] == 0.0):
                data.loc[continent, 'std'] = check_std
            else:
                data.loc[continent, 'std'] = np.std([data.loc[continent, 'std'], check_std]) 
            for key in std_dict:
                std_dict[key] = 0
    for index in data.index:
        data.loc[index,'mean'] = sum(sum_dict[index])/(len(sum_dict[index])+1)
    

    
    return data
        

    
answer_seven()
```
